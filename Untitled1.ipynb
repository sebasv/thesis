```python
# VAR en CVAR functies, en alpha als functie van CVAR (Y,X)
VaR = lambda a: -stats.norm.ppf(a)
CVaR = lambda a: integrate.quad(lambda x: VaR(x), 0, a)[0]/a
X = np.arange(1, 0, -.01)
# plt.plot(X, VaR(X), 'r', X, [CVaR(x) for x in X], 'k');
# print('\n'.join(f'{x:.2f}  {CVaR(x):.4f}' for x in np.arange(.05,1,.05)))
Y = [CVaR(x) for x in X]
```

### Sample price bounds for different gammas. Calculate % within bounds for that value of $\gamma$.

```python
gammas = np.arange(.1, 15, .4)
hits = np.zeros_like(gammas)
for i, gamma in enumerate(gammas):
    S, Mu, Ml = opt.inefficient(W, V, r, .01, .0419, .5193, .1183, gamma, .75)
    hit = 0
    for k in K:
        idx = df.loc[df.K==k, 'T'].mul(100).astype(int).values
        hit += np.sum((df.loc[df.K==k, 'put'].values > P(Ml, k, S)[idx]) & (df.loc[df.K==k, 'put'].values < P(Mu, k, S)[idx]))
    hits[i] += hit/len(df)        
hits
```

### Reconstruct $\sigma_M$ from the values of $\gamma$.

```python
sigma_m = (delta)**2 + gammas**2*ss**2 + (ss*qs/np.mean(V))
sigma_m
```

## The alphas corresponding to the minimum expected shortfall found by reverse engineering $\sigma_m$

```python
alphas = np.interp(sigma_m, Y, X)
alphas
```

```python

def cheyette(kappa, sigma, n, p, W, dt, f):
    X = np.zeros((n+1, p))
    Y = 0
    for i in range(n):
        X[i+1] = X[i] + (Y -kappa*X[i])*dt + sigma*W[i]
        Y     += (sigma**2-2*kappa*Y)*dt
    return X + f

def CIR(x0, W, n, p, dt, kappa, xi, sigma):
    """CIR process from randomness W, initial value x0 and parameters"""
    X = np.ones((n+1, p))*x0
    for i in range(n):
        X[i+1] = np.abs( X[i] + kappa*(xi - X[i])*dt + np.sqrt(X[i])*sigma*W[i] )
    return X


class Calibration(object):
    snu = np.array([1,0,0])
    sV  = np.array([0,1,0])

    def __init__(self, r_xi, r_kappa, r_f, con):
        self.r_params = [r_xi, r_kappa, r_f]
        
        df = pd.read_sql('select * from option_data', con, index_col='index')
        df.columns = df.columns.astype(float)
        self.data = df
        
    def calculate(self, qS, kappa_v, xi_v, sigma_v, delta, sS, gamma=0):
        V = 1 + np.sqrt( CIR(xi_v, self.W[:,:,0], self.discretization_steps, self.n_simulations, self.dt, kappa_v, xi_v, sigma_v)[:-1] )
        beta = -qS / np.inner(sS, sS) / V
        print(np.mean(self.r), qS, self.dt, np.std(self.W), self.sV, np.mean(V))
        S = np.cumprod(1 + (self.r+qS)*self.dt + (self.W @ self.sV) * V , 0)
        p3 = beta * (self.W @ sS) + self.W @ (delta*self.snu)
        lu = self.W @ (gamma * self.sV)
        Mu = np.cumprod(1 -  self.r    *self.dt + p3 + lu, 0)
        Ml = np.cumprod(1 -  self.r    *self.dt + p3 - lu, 0)
        return S, Mu, Ml, V

    def simulate(self, dt=.01, n_simulations=100, T=10, args=None):
        if T < self.data.columns.max():
            raise ValueError("max maturity too small for data")
        self.discretization_steps = int(T/dt)+1
        self.n_simulations = n_simulations
        self.dt = dt
        self.W = np.random.normal(scale=np.sqrt(self.dt), size=(self.discretization_steps, n_simulations, 3))
        self.Wr = np.random.normal(scale=np.sqrt(self.dt), size=(self.discretization_steps-1, n_simulations))
        self.r = cheyette(*self.r_params[:2], *self.Wr.shape, self.Wr, self.dt, self.r_params[2])

        S, Mu, Ml, V = self.calculate(*args[:5], np.r_[args[5:8]], args[8])
        return S, Mu, Ml, V
    
with sqlite3.connect(r'optimization\Optimizer\results.sqlite') as con:
    ca = Calibration(0.0212, 0.00445, .002, con)
    df = pd.read_sql('select * from heston', con)
# S, Mu, Ml = ca.simulate(args=[])
```

```python
from scipy import stats
import numpy as np
def iv_to_call(df):
    d1 = lambda s, k, ss, t: np.log(s/k)/ss/np.sqrt(t) + ss*np.sqrt(t)
    d2 = lambda s, k, ss, t: d1(s,k, ss, t) - ss*np.sqrt(t)
    bs = lambda s, k, ss, t: stats.norm.cdf(d1(s, k, ss, t))*s - stats.norm.cdf(d2(s, k, ss, t))*k
    data = df.copy()
    for k in data.columns:
        for t in data.index:
            data.loc[t, k] = bs(1, k, data.loc[t, k], t) #impvol to BS price
    return data

```
