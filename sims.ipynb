```python
%matplotlib inline
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
N = int(1e6)
W = np.random.normal(size=(2, N))
m, s = 1, np.array([2., 1.])
p = .1
```

```python
Y = np.zeros(shape=(100,2))
for i in range(100):
    S = s/100*i
    X = np.exp(m+np.dot(S, W))
    F = np.maximum(X - np.exp(m + np.inner(S,S)/2), 0)

    ESx = np.mean(X[-X <= np.percentile(-X, 100*p)])
    ESf = np.mean(F[-F <= np.percentile(-F, 100*p)])
    Y[i] = [ESx, ESf]
```

```python
np.mean(X), np.exp(m + np.inner(S, S)/2)
```

```python
plt.plot(Y)
```

```python
f = lambda x,y: np.exp(np.exp(x/y))
plt.plot(f(Y[:,1],Y[:,0]))
```
